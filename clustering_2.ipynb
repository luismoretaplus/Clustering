Segmentación de Clientes:

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')
```

```python
#Observamos que la variable Gender no tiene gran diferenciación
df3 = df2[['Age', 'Annual_Income', 'Spending_Score']]
```

```python
#Vamos a escalar y normalizar los datos
#Esto con el fin de que tengan un rango y varianza similares
scaler = StandardScaler()
df_escalado = scaler.fit_transform(df3)
```

```python
#A través de un grafico de codos, vamos a encontrar el npumero óptimo de klusters que debe tener nuestra segmentación
nc = range(1, 10)
kmeans = [KMeans(n_clusters=i) for i in nc]
score = [kmeans[i].fit(df_escalado).score(df_escalado) for i in range(len(kmeans))]
score
fig, ax = plt.subplots(figsize=(7, 4))
plt.xlabel('Número de clústeres (k)')
plt.ylabel('Suma de los errores cuadráticos')
plt.plot(nc,score)
```

El gráfico indica que el mejor valor de k para los datos es 4.

```python
#Número de grupos (k) a 4.
k = 4
#Establecemos una semilla aleatoria para la reproducibilidad.
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
#Ajustamos los datos escalados
kmeans.fit(df_escalado)
```

```python
#Obtenemos los grupos (clusters) y las etiquetas de los centroides.
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

#Mostramos el gráfico de dispersión.
plt.scatter(df_escalado[:, 0], df_escalado[:, 1], c=labels, s=50, cmap='viridis')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, c='black')
plt.xlabel('longitud')
plt.ylabel('latitud')
plt.title('Centroides obtenidos por K-Means')
plt.show()
```

```python
#Aumentamos a nuestro df original las etiquetas del cluster
df2['cluster'] = labels
```

```python
#Realizamos un analisis de componentes prinmcipales para verificar si los cluster se separan correctamente
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(df_escalado)

plt.scatter(pca_data[:, 0], pca_data[:, 1], c=df2['cluster'], s=50, cmap='viridis')
plt.xlabel('Componente principal 1')
plt.ylabel('Componente principal 2')
plt.title('PCA por cluster')
plt.show()
```

```python
sns.pairplot(df2, hue = 'cluster', palette = 'tab10', diag_kind = 'kde', corner = True)
plt.suptitle('Diagrama de Dispersión por Cluster')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='cluster', y='Annual_Income', data=df2, palette='Set1')
plt.title('Boxplot de Ingreso Anual por Cluster')
plt.xlabel('Cluster')
plt.ylabel('Ingreso Anual')
plt.show()
```
